```python
# setup preprocessor
preprocessor.set_params(
    vectorizer__n_components=256, # dimensionality of the word vectors
    vectorizer__w2v={
        'iter':128, # number of training iterations of the w2v model
        'min_count':2, # all words with total frequency lower than this will be ignored
    },
)
preprocessor.fit(train_data)
train_data = preprocessor.transform(train_data)
test_data = preprocessor.transform(test_data)

train_data.head(10) # see how the preprocessed data looks!
```

```python
def accuracy(y, y_hat) -> float:
    accuracy = metrics.accuracy_score(y, y_hat)
    return accuracy

def f1(y, y_hat, alpha: float = 0.5, beta: float = 1.):
    f1 = metrics.f1_score(y, y_hat, average=None, zero_division=0.)
    return f1

def precision(y, y_hat) -> float:
    precision =  metrics.precision_score(y, y_hat, average=None, zero_division=0.)
    return precision

def recall(y, y_hat) -> float:
    recall = metrics.recall_score(y, y_hat, average=None, zero_division=0.)
    return recall

evaluation_functions = dict(accuracy=accuracy, f1=f1, precision=precision, recall=recall)

def evaluate(y, y_hat) -> th.Dict[str, float]:
    return {name: func(y, y_hat) for name, func in evaluation_functions.items()}
```

```python
class NaiveBayes:

    def __init__(self, kind='gaussian', smoothing=0):
        self.kind = kind
        self.smoothing = smoothing
        self.distribution = []

    def set_type(self, x, y):
        size = x.shape[0]
        if self.kind == 'gaussian':
            self.prior = np.array([x.shape[0] - np.sum(y), np.sum(y)]) / size
            self.m = []
            self.covariance = []
            self.m.append(np.mean(x[y == 0], axis=0))
            self.m.append(np.mean(x[y == 1], axis=0))
            self.covariance.append(x[y == 0])
            self.covariance.append(x[y == 1])
            I0 = np.eye(self.covariance[0].shape[1])
            I1 = np.eye(self.covariance[1].shape[1])
            cov0 = np.cov(np.stack(self.covariance[0], axis=1))
            cov1 = np.cov(np.stack(self.covariance[1], axis=1))
            self.covariance[0] = cov0 + self.smoothing * I0
            self.covariance[1] = cov1 + self.smoothing * I1
        if self.kind == 'bernoulli':
            self.prior = np.array([x.shape[0] - np.sum(y), np.sum(y)], dtype=np.float64) / size
            self.threshold = np.mean(x, axis=0)
            self.p = np.zeros((2, x.shape[1]))

    def fit(self, x, y):
        NaiveBayes.set_type(self, x, y)
        labels = [0, 1]
        if self.kind == 'gaussian':
            for label in labels:
                self.distribution.append(multivariate_normal(self.m[label], self.covariance[label]))
        if self.kind == "bernoulli":
            for label in labels:
                for i in range(x.shape[1]):
                    count = 0
                    to_check = x[y == label][:, i][x[y == label][:, i] > self.threshold[i]]
                    for j in to_check:
                        if j != 0:
                            count += 1
                    self.p[label][i] = (count + self.smoothing) / (x[y == label].shape[0] + 2 * self.smoothing)

    def predict(self, x):
        predictions = []
        for i in range(len(x)):
            prior0 = np.log(self.prior[0])
            prior1 = np.log(self.prior[1])
            if self.kind == "gaussian":
                prior0 += self.distribution[0].logpdf(x[i]).sum()
                prior1 += self.distribution[1].logpdf(x[i]).sum()
            if self.kind == "bernoulli":
                a0 = np.log(self.p[0][x[i] > self.threshold])
                a1 = np.log(self.p[1][x[i] > self.threshold])
                b0 = np.log(1 - self.p[0][x[i] <= self.threshold])
                b1 = np.log(1 - self.p[1][x[i] <= self.threshold])
                prior0 += np.sum(a0) + np.sum(b0)
                prior1 += np.sum(a1) + np.sum(b1)
            if prior0 <= prior1:
                predictions.append(1)
            else:
                predictions.append(0)
        return np.array(predictions)
```

```python
class KNN:
    def __init__(self,k):
        self.k=k

    def fit(self, x, y):
        self.x_train = x
        self.y_train = y
        return self

    def predict(self, x, k):
        size = x.shape[0]
        self.k = k
        squares= np.sum(np.square(self.x_train), axis=1)
        train_transpose=x.dot(self.x_train.T)
        sum = np.sum(np.square(x)[:, np.newaxis, :], axis=2) - 2 * train_transpose  + squares
        cans = np.sqrt(sum)
        predictions = np.zeros(size)
        for i in range(size):
            candidates = [self.y_train[j] for j in np.argsort(cans[i])[0:self.k]]
            predictions[i] = Counter(candidates).most_common(1)[0][0]
        return predictions

    def evaluation(self, x, y, k):
        predictions = self.predict(x, k)
        acc = metrics.accuracy_score(y, predictions)
        return acc.round(5)
```

```python
# param_grid = dict(
    # cluster_count = [potential choices], 
    # max_iters = [potential choices],
    # other hyper-parameters to search over (if any)
# )
# model = clustering.KMeans(
    # hyper-parameters here (if any)
# )
# todo: fit & validate the model to achieve the best purity on training data (vec_1 & vec_2)
# todo: report best settings
# target purity for vec_1>=0.23, and for vec_2>=0.30 

class KMeans():
    
    def __init__(self, x, k):
        self.k = k
        self.x = x
        self.centroids = self.x[np.random.choice(self.x.shape[0], self.k, replace=False)]
        self.classes = np.random.randint(self.k, size=self.x.shape[0])

    def set_class(self):
        for i in range(len(self.classes)):
            for j in range(len(self.centroids)):
                a = self.x[i] - self.centroids[self.classes[i]]
                b = self.x[i] - self.centroids[j]
                if LA.norm(a) >= LA.norm(b):
                    if LA.norm(self.x[i] - self.centroids[j]) != 0:
                        self.classes[i] = j

    def set_center(self):
        self.centroids = np.zeros(self.centroids.shape)
        counts = [0 for j in range(self.k)]
        for i in range(len(self.x)):
            counts[self.classes[i]] += 1
            self.centroids[self.classes[i]] += self.x[i]
        for j in range(len(counts)):
            if counts[j] !=0 :
              self.centroids[j] /= counts[j]
        return self.centroids

    def fit(self, iter):
        for i in range(iter):
            self.set_class()
            self.centroids = self.set_center()

    def predict(self):
        return self.classes
```

Clustring Evaluation

```python
def purity(y, y_hat) -> float:
    contingency_matrix = metrics.cluster.contingency_matrix(y, y_hat)
    purity = np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix)
    return round(purity,5)

def adjusted_rand_index(y, y_hat) -> float:
    return round(adjusted_rand_score(y, y_hat),5)

evaluation_functions = dict(purity=purity, adjusted_rand_index=adjusted_rand_index)

def evaluate(y, y_hat) -> th.Dict[str, float]:
    return {name: func(y, y_hat) for name, func in evaluation_functions.items()}
```
